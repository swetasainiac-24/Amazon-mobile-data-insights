```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.amazon.in/s?k=mobile+phones"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US,en;q=0.5"
}
```

```python
r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, "html.parser")

products = soup.select("div.s-main-slot div[data-component-type='s-search-result']")
```

```python
data = []

for p in products:
    # Name
    try:
        name = p.h2.text.strip()
    except:
        name = "N/A"
```

#price



```python
for p in products:

    # Name
    name_el = p.h2
    name = name_el.get_text(strip=True) if name_el else "N/A"

    # Price
    price_el = p.select_one("span.a-price-whole")
    price = price_el.get_text(strip=True).replace(",", "") if price_el else "N/A"

    # Rating
    try:
        rating_el = p.select_one("span.a-icon-alt")
        rating = rating_el.get_text(strip=True) if rating_el else "N/A"
    except:
        rating = "N/A"

    # Link
    try:
        if p.h2 and p.h2.a:
            link = "https://amazon.in" + p.h2.a.get("href", "")
        else:
            link = "N/A"
    except:
        link = "N/A"

    data.append([name, price, rating, link])

df = pd.DataFrame(data, columns=["Name", "Price", "Rating", "Link"])
df
```
